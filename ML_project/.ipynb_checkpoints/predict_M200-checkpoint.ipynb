#### Random Forest Environment Classification
##### 0 = field, 1 = group/cluster
##### Best result: 76.84% accuracy

```python
#scaffold of the following code is courtesy of ChatGPT, as much time has passed since EECS-658.

#apply the trimming flags!
#note: remove group flag if I'm looking at environment classifications rather than halo mass
nedlvs_parent_test = nedlvs_parent[tempel_flag & wisesize_flag]
```

```python
df = nedlvs_parent_test.to_pandas()
```

```python
#replace -999 with NaNs
df.replace(-999, np.nan, inplace=True)

#REMOVE NaNs
df = df.dropna()
```

```python
#just creating 0, 1, 2 classes based on Tempel flags --> 0 field, 1 group, 2 cluster

df['env_class'] = np.select([df['tempel_field_flag'],df['tempel2017_group_flag']|df['tempel2017_cluster_flag']],
                [0, 1])
```

```python
#predictors...some hodgepodge of potentially relevant properties.
#X = df[['Sigma_M1','Sigma_M2','Sigma_M3','Sigma_M4','Sigma_M5',
#        'Sigma_31','Sigma_51','Sigma_52','Sigma_53']]  
X = df[['Sigma_M1','Sigma_M2','Sigma_M3','Sigma_M4','Sigma_M5','Sigma_M7',
        'Sigma_31','Sigma_33','Sigma_51','Sigma_53']]  

#target variable...environment *class*
y = df['env_class']  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = RandomForestClassifier(class_weight='balanced')
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy = (np.sum(np.diag(confusion_matrix(y_test, y_pred)))) / (np.sum(confusion_matrix(y_test, y_pred))) 
print(f'Accuracy: {accuracy*100:.2f}%')
```

```python
importances = clf.feature_importances_
features = X.columns

plt.barh(features, importances)
plt.xlabel("Feature Importance")
plt.title("Random Forest Feature Importances")
plt.tight_layout()
plt.show()
```

```python
#CORRELATION MAP! Check if there are any redundant features (correlation > 0.95 or so)

import seaborn as sns
corr = X.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

#### Random Forest Regression
##### Using Tempel+2017 groups and M200
##### Best result (all log(M200)): MSE = 0.636 | R² = 0.531 (split random_state = 63; test_size=0.15)
##### Best result (log(M200)>12): MSE = 0.134 | R² = 0.672 (split random_state = 27; test_size=0.3)

```python
#apply the trimming flags!
#note: using halo mass, so must isolate Tempel+2017 group galaxies (which actually have a halo mass)
nedlvs_parent_test = nedlvs_parent[group_flag & tempel_flag & wisesize_flag]
```

```python
nedlvs_parent_test['group_M200'] = np.log10(nedlvs_parent_test['group_M200'])
```

```python
df = nedlvs_parent_test.to_pandas()
```

```python
#replace -999 with NaNs
df.replace(-999, np.nan, inplace=True)

#REMOVE NaNs
df = df.dropna()
```

```python
#predictors...some hodgepodge of potentially relevant properties.
#X = df[['Sigma_M1','Sigma_M2','Sigma_M3','Sigma_M4',
#        'Sigma_31','Sigma_53']][df['group_M200']>12.5]
X = df[['Sigma_M1','Sigma_M3','Sigma_M4','Sigma_M5','Sigma_M6',
        'Sigma_33','Sigma_53','Mstar','SFR_hybrid']]#[df['group_M200']>12.5]

#target variable...log(M200)
y = df['group_M200']#[df['group_M200']>12.5]
```

```python
#split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=63)
```

```python
model = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

# Performance metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse:.3f}")
print(f"R²: {r2:.3f}")
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.5, s=10)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("True log(M200)")
plt.ylabel("Predicted log(M200)")
plt.title("Random Forest Regression")
plt.grid(True)
plt.tight_layout()
plt.show()
```
