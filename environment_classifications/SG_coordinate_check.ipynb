#### Aim One: Use my script with full VFS and Vcosmic from vf_v2_environment.py, compare the outputs to SGX, SGY, SGZ from Gianluca.

```python
#load my RADEC_to_SG function, apply to VFS with redshifts Vcosmic/3e5

import sys
sys.path.append(homedir+'/github/wisesize/')
from universal_functions import RADEC_to_SG

sgx, sgy, sgz = RADEC_to_SG(f['RA'], f['DEC'], fenv['Vcosmic']/3e5)
```

Note how none of the coordinates are exactly 1-to-1 with redshift, though SGY is the closest. This finding alone renders the of SGY as an equivalent redshift "distance" a bit tenuous.

```python
fenv_sgx=fenv['SGX']
fenv_sgy=fenv['SGY']
fenv_sgz=fenv['SGZ']
mask_x = ~np.isnan(fenv_sgx) & ~np.isnan(sgx)
mask_y = ~np.isnan(fenv_sgy) & ~np.isnan(sgy)
mask_z = ~np.isnan(fenv_sgz) & ~np.isnan(sgz)

plt.subplots(1,3, figsize=(20,5))
plt.subplot(1,3,1)
plt.scatter(fenv_sgx/0.74,sgx)
plt.axline([0,0],slope=1)
co_x = np.polyfit(fenv_sgx[mask_x], sgx[mask_x], 1)
print(co_x)
plt.xlabel('My SGX')
plt.ylabel('VFS SGX')

plt.subplot(1,3,2)
plt.scatter(fenv_sgy/0.74,sgy)
plt.axline([0,0],slope=1)
co_y = np.polyfit(fenv_sgy[mask_z], sgy[mask_z], 1)
print(co_y)
plt.xlabel('My SGY')
plt.ylabel('VFS SGY')

plt.subplot(1,3,3)
plt.scatter(fenv_sgz/0.74,sgz)
plt.axline([0,0],slope=1)
co_z = np.polyfit(fenv_sgz[mask_z], sgz[mask_z], 1)
print(co_z)
plt.xlabel('My SGZ')
plt.ylabel('VFS SGZ')

plt.show()
```

A 1-to-1 correlation with Vcosmic (which, as a reminder, is flow-corrected and the like). The reason I multiplied the VFS SGxyz by 1/0.74 is because those SG coordinates are reported in units of h^-1 Mpc, where h = 0.74. 

#### Aim Two: Use my script with full-ish VFS and redshift from NED-LVS, compare the outputs to SGX, SGY, SGZ from Gianluca. Note that there are a few galaxies in VFS missing from NED-LVS, resulting in a marginally smaller sample.

```python
#and now...using NED-LVS redshift
sgx, sgy, sgz = RADEC_to_SG(vfs_nedlvs['RA_1'], vfs_nedlvs['DEC_1'], vfs_nedlvs['z'])
```

```python
vfs_nedlvs_sgx=vfs_nedlvs['SGX']
vfs_nedlvs_sgy=vfs_nedlvs['SGY']
vfs_nedlvs_sgz=vfs_nedlvs['SGZ']
mask_x = ~np.isnan(vfs_nedlvs_sgx) & ~np.isnan(sgx)
mask_y = ~np.isnan(vfs_nedlvs_sgy) & ~np.isnan(sgy)
mask_z = ~np.isnan(vfs_nedlvs_sgz) & ~np.isnan(sgz)

plt.subplots(1,3, figsize=(20,5))
plt.subplot(1,3,1)
plt.scatter(vfs_nedlvs_sgx/0.74,sgx)
plt.axline([0,0],slope=1)
co_x = np.polyfit(vfs_nedlvs_sgx[mask_x], sgx[mask_x], 1)
plt.ylabel('NED-LVS SGX')
plt.xlabel('VFS SGX')
print(co_x)

plt.subplot(1,3,2)
plt.scatter(vfs_nedlvs_sgy/0.74,sgy)
plt.axline([0,0],slope=1)
plt.ylabel('NED-LVS SGY')
plt.xlabel('VFS SGY')
co_y = np.polyfit(vfs_nedlvs_sgy[mask_z], sgy[mask_z], 1)
print(co_y)

plt.subplot(1,3,3)
plt.scatter(vfs_nedlvs_sgz/0.74,sgz)
plt.axline([0,0],slope=1)
plt.ylabel('NED-LVS SGZ')
plt.xlabel('VFS SGZ')
co_z = np.polyfit(vfs_nedlvs_sgz[mask_z], sgz[mask_z], 1)
print(co_z)

plt.show()
```

Not nearly as well correlated as when I used Vcosmic, perhaps because these redshifts are direct measurements and do not account for flow corrections, etc. Indeed, the main loci for each plot seem to follow a shallower slope than what the 1-to-1 line traces.

Anyhow, I will be abandoning SG coordinates in favor of RA-DEC spherical trigonometry to measure two-dimensional distances between objects. Tah-tah.
